```python
import numpy as np
import pandas as pd
import math
from scipy import stats
```

```python
data = pd.read_csv("./data-knn/data.csv")
data = data.drop(columns = ["Unnamed: 32", "id"])
data['diagnosis'] = data['diagnosis'].map({"M": 1, "B": 0})


frac_test = 0.8
split_idx = int(frac_test * data.shape[0])
y_col = 'diagnosis'


# shuffling and splitting
data = data.sample(frac = 1, random_state=1) 
data_train = data[: split_idx]
data_test = data[split_idx: ]
data_train.head()
```

```python
X_train = data_train.drop(columns = y_col).values
y_train = data_train[y_col].values

X_test = data_test.drop(columns = y_col).values
y_test = data_test[y_col].values

X_train.shape, X_test.shape
```

```python
# assuming euclidien distance only for now
# init -- k
# fit -- X_train, y_train
# predict -- X_test

class KNN():
    def  __init__(self, n_neighbors: int = 5):
        if not isinstance(n_neighbors, int) or n_neighbors <= 0:
            raise ValueError("n_neighbors must be a positive integer")
        self.k = n_neighbors

    def fit(self, X_train: np.array, y_train: np.array):
        assert X_train.shape[0] == y_train.shape[0], "shape mismatch in the training samples"
        self.X_train = X_train
        self.y_train = y_train

    def predict(self, X_test):
        # for every element in X_test
        # we calculate the distance between every point in training data
        # Sort the distances in decreasing order
        # Select the labels corresponding to top-k training points
        # Predict the most-occuring label as its label

        assert self.X_train is not None or self.y_train is not None, "Model not fitted. Call fit() before predict()"

        predictions = []
        for test_sample in X_test:
            distances = [self._calculate_dist(test_sample, x) for x in self.X_train]
            top_indices = np.argsort(distances)
            top_k_labels = self.y_train[top_indices][:self.k]
            prediction = stats.mode(top_k_labels).mode 
            prediction = int(prediction)
            predictions.append(prediction)
            
        return predictions

    def _calculate_dist(self, x1, x2):
        return math.sqrt(np.sum((x1 - x2)**2))
    
```

```python
knn_sahil = KNN(n_neighbors=20)
# knn_sahil
knn_sahil.fit(X_train=X_train, y_train=y_train)
knn_sahil._calculate_dist(X_train[0], X_test[14])
knn_sahil.predict(X_test=X_test[:10])
```
